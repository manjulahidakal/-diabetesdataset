```python
import pandas as pd
import numpy as np

df = pd.read_csv("diabetes.csv")
```

```python
df.columns = df.columns.str.strip().str.lower()
```

```python
# columns where 0 is not a valid medical value

invalid_zero_cols = [
  'glucose',
  'bloodpressure',
  'skinthickness',
  'insulin',
  'bmi'
]

# replace 0 with nan
df[invalid_zero_cols] = df[invalid_zero_cols].replace(0, np.nan)
```

```python
df.isnull().sum()
# (missing value check code)
```

```python
# impute missing values

for col in invalid_zero_cols:
  df[col].fillna(df[col].median(),inplace=True)
```

```python
# median represents: a typicl patient not too high not too low not influence bu outlier
```

```python
# feature target separetion

# feature
x = df.drop(columns = ['outcome'])

# target
y = df['outcome']
```

```python
df.columns
```

```python
x.columns
```

```python
y.name
```

```python
# train and test split

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
x_train.shape,x_test.shape
```

```python
y.value_counts(normalize=True)
```

```python
y_train.value_counts(normalize=True)            #stratify working as no we use  #normalize=True gives percentages instead of counts
```

```python
y_test.value_counts(normalize=True)
```

```python
#feature scaling

from sklearn.preprocessing import StandardScaler
scalar = StandardScaler ()

x_train_scaled = scalar.fit_transform(x_train)
x_test_scaled = scalar.transform(x_test)
```

```python
# here using logistic regression classifier because it is a binary classification problem, strong baseline for medical data

from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(x_train_scaled,y_train)

# here now the model trained, uses scaled features , can implement safely
```

```python
#predictions

y_pred = log_reg.predict(x_test_scaled)
```

```python
y_pred[:10]
```

```python
log_reg.score(x_test_scaled,y_test)

# accuary of test
```

```python
# accuary of train

log_reg.score(x_train_scaled,y_train)
```

```python
# fit_transform means learn from the data and change it at the same time.
                           or
# - Fit + Transform together  fit_transform, usually used on training data.
```

```python
y_prob = log_reg.predict_proba(x_test_scaled)
```

```python
y_prob[:20]
```

```python
y_prob_diabetic = y_prob[:,1]
y_prob_diabetic[:10]
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
#[TN FP]
# [FN TP]

#  TN- true negative  - actually not diabetic
#  FP- false positive  - actually non diabetic but predicted diabetic
#  FN- false negative   - actually diabetic but predicted non diabetic
#  TP- true positive   - actually diabetic
```

```python
# classification report

from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))

# here we will see about precision, recall, f1 score and support
# precision: how correct the model was when it said 0 or 1
# recall: how many real 0 and 1 it found
# f1:balance score of precision and recall
# support : how many actual 0 and 1 patients

# example for weighted avg
# classes with more samples influence the final score more than the smaller classes
```

```python
#next model we using here is SVM Support Vector Machine ,
# it finds a strong , margin based decision
#boundary that  often imporves recall on small ,
# well scalled dataset

# the main reason to use is dataset is small in size,
# we did the scalling , chances are high to get better
# recall
```

```python
from sklearn.svm import SVC
svm_model = SVC(kernel ='linear', probability=True)
svm_model.fit(x_train_scaled,y_train)
```

```python
y_pred_svm = svm_model.predict(x_test_scaled)

# evaluating
svm_model.score(x_test_scaled,y_test)
```

```python
y_prob_svm = svm_model.predict_proba(x_test_scaled)[:,1]
```

```python
y_prob_svm[:10]
#
```

```python
# confusion matrix

from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(y_test,y_pred_svm))
```

```python
# ROC AUC Receiver Operating Characteristic, Area Under the Curve

from sklearn.metrics import roc_auc_score
roc_auc_lr = roc_auc_score(y_test,y_prob[:,1])
roc_auc_svm = roc_auc_score(y_test,y_prob_svm)

roc_auc_lr,roc_auc_svm

# 81 for both lr and svm
# roc implies that both LR and SVM correctly ranks a diabetics higher
# then a non diabetic 82% of the time
# slides the threshold from 0 - 1, computes recall and false positives
# at each step, plots the curve
```
